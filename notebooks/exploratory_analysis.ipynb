```python
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir("C:/Users/DELL/OneDrive/Desktop/MOD_sim/Simulation")
```

# Charger tous les résultats

```python
df_mm1 = pd.read_csv("data/mm1_results.csv")
df_gm1 = pd.read_csv("data/gm1_results.csv")
df_mg1 = pd.read_csv("data/mg1_results.csv")
```

# Comparaison des temps de réponse

```python
plt.figure(figsize=(9, 5))
plt.plot(df_mm1["lambda"], df_mm1["avg_response_time"], label="M/M/1", marker='o')
plt.plot(df_gm1["lambda"], df_gm1["avg_response_time"], label="G/M/1", marker='s')
plt.plot(df_mg1["lambda"], df_mg1["avg_response_time"], label="M/G/1", marker='^')
plt.title("Comparaison du temps de réponse moyen")
plt.xlabel("λ")
plt.ylabel("Temps de réponse moyen")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Comparaison des temps d'attente

```python
plt.figure(figsize=(9, 5))
plt.plot(df_mm1["lambda"], df_mm1["avg_waiting_time"], label="M/M/1", marker='o')
plt.plot(df_gm1["lambda"], df_gm1["avg_waiting_time"], label="G/M/1", marker='s')
plt.plot(df_mg1["lambda"], df_mg1["avg_waiting_time"], label="M/G/1", marker='^')
plt.title("Comparaison du temps d'attente moyen")
plt.xlabel("λ")
plt.ylabel("Temps d'attente moyen")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Évolution de l'utilisation 

```python
plt.figure(figsize=(9, 5))
plt.plot(df_mm1["lambda"], df_mm1["utilization"], label="M/M/1", marker='o')
plt.plot(df_gm1["lambda"], df_gm1["utilization"], label="G/M/1", marker='s')
plt.plot(df_mg1["lambda"], df_mg1["utilization"], label="M/G/1", marker='^')
plt.title("Utilisation du serveur selon λ")
plt.xlabel("λ")
plt.ylabel("Utilisation (ρ)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
